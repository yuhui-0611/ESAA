3. 변수를 변환합니다.

```python
# type에는 white와 red 두 종류가 있습니다.
# 각각 0,1로 변환합니다.
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
train['type'] = encoder.fit_transform(train['type'])
test['type'] = encoder.transform(test['type'])
```

```python
# 모델에 입력하기 전 데이터를 정형합니다.
# train_x : train의 'index', 'quality' 피처를 제거하고 저장합니다.
# train_y : train의 'quality'를 저장합니다.
# test_x : test의 'index'를 제거하고 저장합니다.

train_x = train.drop(['index','quality'],axis = 1)
train_y = train['quality']
test_x = test.drop(['index'],axis = 1)
```

4. LightGBM 모델을 생성 및 훈련합니다.

```python
lgbm_model = lgbm.LGBMClassifier()
lgbm_model.fit(train_x,train_y)
```

5. 생성한 모델로 예측 데이터를 생성합니다.

```python
pred = lgbm_model.predict(test_x)
```

6. 제출파일을 생성합니다.

```python
smpl_sub.head()
```

```python
#서브미션 파일의 quality 변수에 예측값을 대입합니다.
smpl_sub['quality'] = pred
```

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from xgboost import XGBClassifier

train = pd.read_csv('/content/drive/MyDrive/ESAA/train.wine.csv')
test = pd.read_csv('/content/drive/MyDrive/ESAA/test.wine.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())
import pandas as pd
train = pd.get_dummies(train, columns=['type'],prefix=['type'])
test = pd.get_dummies(test, columns=['type'],prefix=['type'])
```

```python
# 모델 정의
xgb_model = XGBClassifier()
```

```python
# 모델 학습
# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수

X = train.drop('quality',axis = 1)
y = train['quality']

y=y-y.min()

# fit 메소드를 이용해 모델 학습

xgb_model.fit(X,y)
```

```python
# predict 메소드와 원핫 인코딩한 test 데이터를 이용해 품질 예측
pred = xgb_model.predict(test)
```

```python
# sample_submission.csv 파일을 불러와 예측된 값으로 채워 주기
sample_sub = pd.read_csv('data/sample_submission.csv')
sample_sub['quality'] = pred
```

```python
# 정답파일 내보내기
sample_sub.to_csv('xgb_pred.csv',index = False)
```

# 3. 보팅

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

train = pd.read_csv('/content/drive/MyDrive/ESAA/train.wine.csv')
test = pd.read_csv('/content/drive/MyDrive/ESAA/test.wine.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())
train = pd.get_dummies(train, columns=['type'], prefix=['type'])
test = pd.get_dummies(test, columns=['type'], prefix=['type'])
```

```python
# 모델 정의 (LightGBM, XGBoost, Random Forest)
lgbm_clf = LGBMClassifier()
xgb_clf = XGBClassifier()
rf_clf = RandomForestClassifier()

# VotingClassifier 정의
voting_clf = VotingClassifier(
    estimators=[('lgbm', lgbm_clf), ('xgb', xgb_clf), ('rf', rf_clf)]
)
```

```python
# 모델 학습
# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수
X_train = train.drop('quality', axis=1)
y_train = train['quality']

# fit 메소드를 이용해 모델 학습
voting_clf.fit(X_train, y_train)
```

```python
# predict 메소드와 원핫 인코딩한 테스트 데이터를 이용해 품질 예측
from sklearn import metrics
from sklearn.metrics import accuracy_score

pred = voting_clf.predict(X_train)
print('Voting 분류기 정확도: {0:.4f}'.format(accuracy_score(y_train, pred)))
```
