## 코드 흐름 요약

- 텍스트 데이터에 대한 전처리 및 벡터화 후 K-Means 클러스터링을 수행하는 과정

1. 텍스트 분석을 위해 필요한 라이브러리들 불러오기
  - nltk, pandas, sklearn 등의 라이브러리를 사용하여 자연어 처리와 머신러닝 분석을 수행
  - 이후 nltk.download()를 이용하여 토큰화와 표제어 추출에 필요한 punkt, punkt_tab, wordnet 데이터를 다운로드

2. LemNormalize 함수를 정의하여 텍스트 데이터를 토큰화(tokenization)하고 표제어 추출(lemmatization)을 수행
  - 이를 통해 문장의 단어들을 기본형으로 변환하여 텍스트 분석의 정확도를 높임

3. 반복문을 통해 각 파일을 pandas 데이터프레임으로 읽어온 뒤, 파일 이름과 텍스트 내용을 각각 리스트에 저장

4. TfidfVectorizer를 이용하여 텍스트 데이터를 TF-IDF 벡터 형태로 변환
  - 이 과정에서 tokenizer=LemNormalize, stop_words='english', ngram_range=(1,2) 등의 옵션을 통해 텍스트 특징을 추출

5. 변환된 벡터 데이터를 KMeans 알고리즘에 적용하여 문서들을 3개의 클러스터로 분류
  - 클러스터링 결과는 각 문서에 대해 cluster_label로 저장되어 데이터프레임에 추가


> **텍스트 데이터 로드 → 텍스트 전처리 → TF-IDF 벡터화 → K-Means 클러스터링 → 클러스터 결과 저장**

## 새롭게 알게 된 내용 / 어려운 내용 / 배울 점

- 자연어 처리에서 텍스트 데이터를 분석하기 위해 필요한 전처리 과정과 벡터화 과정에 대해 이해할 수 있었음
- 특히 NLTK를 이용하여 텍스트 데이터를 토큰화하고 표제어 추출을 통해 단어의 기본형을 사용하는 과정이 매우 중요
  - 이러한 전처리 과정은 텍스트 분석의 정확도를 높이고 의미적으로 유사한 단어들을 동일한 형태로 처리하는 데 도움이 됨
- TF-IDF(Term Frequency-Inverse Document Frequency)는 단순히 단어의 빈도만 고려하는 것이 아니라, 전체 문서에서 얼마나 희귀한 단어인지까지 고려하여 단어의 중요도를 계산
  - 이를 통해 텍스트 데이터를 숫자 벡터 형태로 변환하여 머신러닝 모델에 입력할 수 있다는 것을 이해하였다.

- lambda 함수와 re.sub()를 이용한 정규표현식 기반 텍스트 전처리
  - \d+와 같은 정규표현식: 숫자를 의미
  - 정규표현식은 텍스트 데이터에서 불필요한 정보를 제거하는 데 매우 유용
- glob.glob() = 특정 경로에 있는 여러 파일을 자동으로 읽어오는 방식
- 이를 통해 많은 텍스트 파일을 효율적으로 처리할 수 있음


< 텍스트 전처리 및 숫자 제거 코드 >
```
train_df['document'] = train_df['document'].apply(
    lambda x: re.sub(r'\d+', ' ', x)
)
```


< TF-IDF 벡터화 코드 >
```
tfidf_vect = TfidfVectorizer(tokenizer=LemNormalize,
                             stop_words='english',
                             ngram_range=(1, 2),
                             min_df=0.05,
                             max_df=0.85)

feature_vect = tfidf_vect.fit_transform(document_df['opinion_text'])
```


< K-Means 클러스터링 코드 >
```
km_cluster = KMeans(n_clusters=3, max_iter=1000, random_state=0)
km_cluster.fit(feature_vect)
document_df['cluster_label'] = km_cluster.labels_
```
