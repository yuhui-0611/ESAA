## 1. 추천 시스템의 개요와 배경

### **추천 시스템의 개요**

- 추천 시스템은 전자상거래와 콘텐츠 플랫폼에서 핵심 경쟁력으로 작용함
- 아마존, 유튜브, 애플 뮤직 등은 추천 시스템을 통해 사용자의 취향을 파악하고 맞춤형 상품과 콘텐츠를 제공하여 이용 시간을 늘리고 매출을 높이고 있음
- 추천 엔진은 단순히 사용자가 좋아할 만한 것을 보여주는 데 그치지 않고, 사용자가 몰랐던 새로운 취향까지 발견하게 해주는 역할을 함
- 몇 개의 콘텐츠만 선택해도 관련된 음악, 영화, 영상이 연속적으로 추천되며, 이러한 과정이 사용자를 플랫폼에 더 오래 머무르게 만듦
- 전자상거래 분야에서도 추천 시스템의 효과는 매우 큼
- 아마존, 이베이 등은 추천 시스템 도입 후 매출이 크게 증가하였으며, 추천의 품질은 사이트의 평판과 사용자 만족도를 좌우하는 중요한 요소가 됨
- 사용자의 관심사에 맞는 상품을 정확히 제시할수록 쇼핑의 즐거움도 커짐
- 추천 시스템의 가장 큰 장점은 사용자의 숨은 취향을 발견하고, 그에 맞는 새로운 콘텐츠를 제안해준다는 점임
- 사용자는 이러한 경험을 통해 플랫폼을 더욱 신뢰하게 되고, 더 많은 추천 콘텐츠를 소비하게 됨
- 그 결과 더 많은 데이터가 축적되어 추천이 더욱 정교해지는 선순환 구조가 형성됨
- 결국 경쟁이 치열한 환경 속에서 전자상거래 업체와 영화·음악 콘텐츠 제공 업체들은 사용자의 관심을 오래 유지하기 위해 추천 시스템 고도화에 지속적으로 비용과 노력을 투자하고 있음

### **온라인 스토어의 필수 요소, 추천 시스템**

- 추천 시스템은 특히 온라인 환경에서 큰 가치를 발휘함
- 전자상거래 업체와 온라인 콘텐츠 제공 업체는 매우 많은 상품과 콘텐츠를 보유하고 있기 때문에, 사용자는 선택 과정에서 어려움과 부담을 느끼기 쉬움
- 이러한 문제를 해결하는 핵심 수단이 추천 시스템임
- 좋은 추천 시스템은 사용자가 원하는 상품이나 콘텐츠를 빠르게 찾아주어 온라인 쇼핑과 이용 경험의 즐거움을 높여 줌
- 따라서 추천 시스템은 온라인 스토어에서 필수적인 구성 요소라고 볼 수 있음
- 추천 시스템은 온라인 스토어가 보유한 다양한 고객·상품 관련 데이터를 바탕으로 작동함
- 예를 들어 구매 이력, 둘러본 상품, 장바구니에 넣은 상품, 사용자의 평점, 사용자가 직접 작성한 취향 정보, 클릭한 항목 등이 추천 시스템의 주요 데이터로 활용됨
- 이러한 데이터를 기반으로 추천 시스템은 ‘당신만을 위한 최신 상품’, ‘이 상품을 선택한 다른 사람들이 좋아하는 상품들’과 같은 방식으로 사용자 맞춤형 추천을 제공함
- 즉, 추천 시스템은 방대한 정보 속에서 사용자가 관심을 가질 만한 대상을 선별하여 구매와 선택을 유도하는 중요한 역할을 함

## 2. 콘텐츠 기반 필터링 추천 시스템

- 콘텐츠 기반 필터링은 사용자가 좋아한 아이템과 내용이 비슷한 다른 아이템을 추천하는 방식임
  - 이 방식은 사용자가 높게 평가한 아이템의 장르, 감독, 배우, 키워드 같은 콘텐츠 정보를 분석하여 유사한 대상을 추천함
  - 즉, 사용자 취향을 아이템의 속성으로 파악해 비슷한 콘텐츠를 연결해주는 방식임
  - 예시로는 사용자가 높은 점수를 준 영화의 장르와 감독, 출연 배우, 키워드 등을 바탕으로 비슷한 영화가 추천되는 경우가 있음
> 따라서 콘텐츠 기반 필터링의 핵심은 사용자와 비슷한 다른 사람이 아니라, 사용자가 선호한 아이템 자체의 특징을 기준으로 추천한다는 점임

## 3. 최근접 이웃 협업 필터링

image.png

- 협업 필터링은 비슷한 취향이나 행동을 보인 다른 사용자들의 정보를 바탕으로 아이템을 추천하는 방식임
  - 이 방식은 아이템의 내용 자체보다 사용자 평점, 구매 이력 같은 사용자 행동 데이터에 의존함
  - 즉, 사용자가 아직 평가하지 않은 아이템에 대해 예측 평점을 계산하여 추천하는 것이 핵심임
- 협업 필터링의 주요 목표는 사용자-아이템 평점 매트릭스를 기반으로 ***사용자가 평가하지 않은 아이템의 선호도를 예측***하는 것임
- 이때 행은 사용자, 열은 아이템, 각 값은 평점으로 구성됨
- 또한 협업 필터링은 **최근접 이웃 방식**과 **잠재 요인 방식**으로 나뉘며, 두 방식 모두 사용자-아이템 평점 행렬 데이터를 기반으로 추천을 수행함

image.png

- 사용자-아이템 평점 행렬은 보통 대부분의 값이 비어 있는 희소 행렬 형태를 가짐
  - 이는 사용자가 모든 아이템에 평점을 매기지 않기 때문임
- 최근접 이웃 협업 필터링은 메모리 기반 협업 필터링이라고도 하며, **사용자 기반 방식**과 **아이템 기반 방식**으로 나뉨
  - 사용자 기반 방식은 특정 사용자와 비슷한 취향을 가진 다른 사용자들을 찾아, 그들이 좋아한 아이템을 추천하는 방식임
  - 아이템 기반 방식은 특정 아이템과 함께 자주 선택된 다른 아이템을 추천하는 방식임
> 즉, 최근접 이웃 협업 필터링의 핵심은 사용자 또는 아이템 간 유사도를 계산하고, 가장 비슷한 대상들을 뽑아 추천에 활용하는 것임

- 최근접 이웃 협업 필터링은 대부분 아이템 기반의 알고리즘을 적용
- 유사도 측정 방법인 코사인 유사도는 추천 시스템의 유사도 측정에 가장 많이 적용
  - 추천 시스템에 사용되는 데이터는 피처 벡터화된 텍스트 데이터와 동일하게 다차원 희소 행렬이라는 특징이 있으므로 유사도 측정을 위해 주로 코사인 유사도를 이용
    - 절대적인 값의 크기보다 선호 패턴의 유사성 이 더 중요하기 때문임
    - 코사인 유사도는 벡터의 크기보다 방향의 유사성 을 측정하므로 이러한 데이터에 잘 맞기 때문임
    - 사용자의 점수 습관 차이, 문서 길이 차이 같은 크기 차이의 영향은 줄이고, 실제 취향 패턴의 유사성은 잘 반영하기 때문임

## 4. 잠재 요인 협업 필터링

- 잠재 요인 협업 필터링은 사용자-아이템 평점 행렬 속에 숨어 있는 잠재 요인을 추출해 추천 예측을 할 수 있게 하는 기법
- 대규모 다차원 행렬을 SVD와 같은 차워 감소 기법으로 분해하는 과정에서 잠재 요인을 추출
  - 이러한 기법 = 행렬 분해(Matrix Factorization)
- 다차원 희소 행렬인 사용자-아이템 행렬 데이터를 아래 두 가지로 분해 가능
  - 저차원 밀집 행렬의 사용자-잠재 요인 행렬
  - 아이템-잠재 요인 행렬의 전치 행렬(즉, 잠재 요인-아이템 행렬)
- 이렇게 분해된 두 행렬의 내적을 통해 새로운 예측 사용자—아이템 평점 행렬 데이터를 만들어서 사용자가 아직 평점을 부여하지 않는 아이템에 대한 예측 평점을 생성하는 것
- 이러한 행렬 분해 기법을 이용해 사용자-잠재 요인 행렬과 아이템-잠재 요인 행렬의 전치 행렬로 분해된 데이터 세트를 다시 내적 곱으로 결합하면서 사용자가 예측하지 않은 아이템에 대한 평점을 도출

### **확률적 경사 하강법을 이용한 행렬 분해**

- SVD는 null이 없는 행렬에만 적용 가능
- 따라서 확률적 경사 하강법(= SGD) 이용
  - P와 Q 행렬로 계산된 예측 R 행렬 값이 실제 R 행렬 값과 가장 최소의 오류를 가질 수 있도록 반복적인 비용 함수 최적화를 통해 P와 Q를 유추하는 것

1. P와 Q를 임의의 값을 가진 행렬로 설정
2. P와 Q.T 값을 곱해 예측 R 행렬을 계산하고 예측 R 행렬과 실제 R 행렬에 해당하는 오류 값을 계산
3. 이 오류 값을 최소화할 수 있도록 P와 Q 행렬을 적절한 값으로 각각 업데이트
4. 만족할 만한 오류 값을 가질 때까지 2, 3번 작업을 반복하면서 P와 Q 값을 업데이트해 근사화

> L2 규제를 반영해 실제 R 행렬 값과 예측 R 행렬 값의 차이를 최소화하는 방향성을 가지고 P행렬과 Q행렬에 업데이트 값을 반복적으로 수행하면서 최적화된 예측 R 행렬을 구하는 방식이 SGD 기반의 행렬 분해

```python
import numpy as np

# 원본 행렬 R 생성, 분해 행렬 P와 Q 초기화, 잠재 요인 차원 K는 3으로 설정
R = np.array([[4, np.nan, np.nan, 2, np.nan],
              [np.nan, 5, np.nan, 3, 1],
              [np.nan, np.nan, 3, 4, 4],
              [5, 2, 1, 2, np.nan]])
num_users, num_items = R.shape
K=3

# P와 Q 행렬의 크기를 지정하고 정규 분포를 가진 임의의 값으로 입력
np.random.seed(1)
P = np.random.normal(scale=1./K, size=(num_users, K))
Q = np.random.normal(scale=1./K, size=(num_items, K))
```

```python
from sklearn.metrics import mean_squared_error

def get_rmse(R, P, Q, non_zeros):
  error=0
  # 두 개의 분해된 행렬 P와 Q.T의 내적으로 예측 R 행렬 생성
  full_pred_matrix = np.dot(P, Q.T)

  # 실제 R 행렬에서 널이 아닌 값의 위치 인덱스 추출해 실제 R 행렬과 예측 행렬의 RMSE 추출
  x_non_zero_ind = [non_zero[0] for non_zero in non_zeros]
  y_non_zero_ind = [non_zero[1] for non_zero in non_zeros]
  R_non_zeros = R[x_non_zero_ind, y_non_zero_ind]
  full_pred_matrix_non_zeros = full_pred_matrix[x_non_zero_ind, y_non_zero_ind]
  mse = mean_squared_error(R_non_zeros, full_pred_matrix_non_zeros)
  rmse = np.sqrt(mse)

  return rmse
```

```python
# R > 0인 행 위치, 열 위치, 값을 non_zeros 리스트에 저장
non_zeros = [ (i, j, R[i, j]) for i in range(num_users) for j in range(num_items) if R[i, j] > 0 ]

steps=1000
learning_rate=0.01
r_lambda=0.01

# SGD 기법으로 우와 Q 매트릭스를 계속 업데이트
for step in range(steps):
  for i, j, r in non_zeros:
    # 실제 값과 예측 값의 차이인 오류 값 구함
    eij = r - np.dot(P[i, :], Q[j, :].T)
    # Regularization을 반영한 SGD 업데이트 공식 적용
    P[i, :] = P[i, :] + learning_rate * (eij * Q[j, :] - r_lambda * P[i, :])
    Q[j, :] = Q[j, :] + learning_rate * (eij * P[i, :] - r_lambda * Q[j, :])

    rmse = get_rmse(R, P, Q, non_zeros)
    if (step % 50) == 0 :
      print("### iteration step : ", step, " rmse : ", rmse)
```

```python
pred_matrix = np.dot(P, Q.T)
print('예측 행렬:\n', np.round(pred_matrix, 3))
```
