# 파이토치 기초

< 파이토치 특징 >

**GPU에서 텐서 조작 및 동적 신경망 구축이 가능한 프레임워크**

- GPU
  - 빠른 연산 가능
  - CUDA, cuDNN이라는 API
- 텐서
  - 파이토치 데이터 형태
  - 단일 데이터 형식으로 된 자료들의 다차원 행렬
  - 간단한 명령어(변수 뒤에 .cuda() 추가)를 사용해 GPU로 연산 수행 가능
    -  1 차원 축(행)=axis 0=벡터
    - 2차원 축(열)=axis 1 =행렬
    - 3차원 축(채널)=axis 2=텐서
    > torch.tensor()
- 동적 신경망
  - 훈련을 반복할 때마다 네트워크 변경이 가능한 신경망
  - 학습 중 은닉층을 추가하거나 제거하는 등 모델의 네트워크 조작 가능

## 파이토치 API

- **torch**: GPU를 지원하는 텐서 패키지로, 다차원 텐서 기반 수학 연산을 빠르게 수행
- **torch.autograd**: 자동 미분을 통해 연산 그래프를 즉시 계산하고, 신경망 수정 사항을 효율적으로 반영
- **torch.nn**: 신경망을 쉽게 구성하고 학습할 수 있게 해주는 패키지이며, 합성곱 신경망·순환 신경망·정규화 등을 포함
- **torch.multiprocessing**: 프로세스 전반에 걸쳐 텐서 메모리 공유를 가능하게 하여, 여러 프로세스가 동일한 텐서에 접근·사용할 수 있게 함
- **torch.utils**: DataLoader를 포함해 데이터 제공, 디버깅, 체크포인트 등 다양한 유틸리티 기능을 제공

## 연산 처리

- C/CUDA 패키지가 상위 API에서 할당된 대부분의 계산을 수행
- CPU와 GPU를 이용해 효율적인 데이터 구조와 다차원 텐서 연산을 처리
- 파이토치에서 텐서가 핵심적인 역할을 함
- 텐서는 메모리에 저장될 때 항상 1차원 배열 형태로 저장됨
- 이 1차원 배열 형태의 저장 구조 = 스토리지

< 스토리지 >
- **오프셋**
  - 텐서의 첫 번째 원소가 스토리지에 저장된 시작 인덱스
- **스트라이드**
  - 각 차원에서 다음 원소로 이동하기 위해 건너뛰는 스토리지 원소 수
  - 같은 값을 가진 텐서라도 shape이 다르면 서로 다른 텐서로 다뤄짐
  - A와 Aᵀ는 스토리지 값은 같아도 shape과 stride가 다름
  image.png

- 위 그림 예
  - 스토리지에서 2를 얻기 위해서는
    - A(2×3)의 1에서, 다음 열로 1칸을, 4를 얻기 위해서, 다음 행으로 3칸을 뛰어넘어야 함
      - 따라서 텐서에 대한 스토리지의 stride는 (3, 1) = (다음 행, 다음 열)
    - Aᵀ(3×2)의 stride는 (2, 1)
- 오프셋과 스트라이드는 데이터 자체가 아니라 행렬·텐서의 구조를 구분하는 데 사용

## 모델 정의

- 계층

< nn.Module()을 상속하여 정의하는 방법 >
```
import torch.nn as nn

class MLP(nn.Module):
    def __init__(self, inputs):
        super(MLP, self). __init__()
        self.layer = nn.Linear(inputs, 1) # 계층 정의
        self.activation = nn.Sigmoid() # 활성화 함수 정의

    def forward(self, X):
        X = self.layer(X)
        X = self.activation(X)
        return X
```



< Sequential 신경망을 정의하는 방법 >
- nn.Sequential을 사용하면 __ init __ ()에서 사용할 네트워크 모델들을 정의해 줄 뿐만 아니라 forward( ) 함수에서는 모댈에서 실행되어야 할 계산을 좀 더 가독성이 뛰어나게 코드로 작성 가능
- Sequential 객체는 그 안에 포함된 각 모률을 순차적으로 실행 가능
- nn.Sequential은 모델의 계층이 복잡할수록 효과가 good

````python
import torch.nn as nn

class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=64, kernel_size=5),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2))

        self.layer2 = nn.Sequential(
            nn.Conv2d(in_channels=64, out_channels=30, kernel_size=5),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2))

        self.layer3 = nn.Sequential(
            nn.Linear(in_features=30*5*5, out_features=10, bias=True),
            nn.ReLU(inplace=True))

    def forward(self, x):
        x = self.layer1(x)
        x = self.layer2(x)
        x = x.view(x.shape[0], -1)
        x = self.layer3(x)
        return x

model = MLP() # 모델에 대한 객체 생성

print("Printing children\n-----------------------------------------")
print(list(model.children()))

print("\n\nPrinting Modules\n-----------------------------------------")
print(list(model.modules()))
````

````python
def MLP(in_features=1, hidden_features=20, out_features=1):
    hidden = nn.Linear(
        in_features=in_features,
        out_features=hidden_features,
        bias=True
    )
    activation = nn.ReLU()
    output = nn.Linear(
        in_features=hidden_features,
        out_features=out_features,
        bias=True
    )

    net = nn.Sequential(hidden, activation, output)
    return net
````

## 모델의 파라미터 정의

- 손실 함수
  - 학습하는 동안 출력과 실제 값(정답) 사이의 오차 측정
  - BCELoss : 이진 분류
  - CrossEntropyLoss : 다중 클래스 분류
  - MSELoss : 회귀 모델
- 옵티마이저
  - 데이터와 손실 함수를 바탕으로 모델 업데이트 방법 결정
  - 특성
    - step() 메서드를 통해 전달받은 파라미터 업데이트
    - 모델의 파라미터별로 다른 기준 적용 가능
    - torch.optim.Optimizer(params, defaults)는 모든 옵티마이저의 기본이 되는 클래스
    - zero_grad() 메서드는 옵티마이저에 사용된 파라미터들의 기울기를 0으로 만듦
    - torch.optim.lr_scheduler는 에포크에 따라 학습률 조절 가능
- 학습률 스케쥴러
  - 미리 지정한 횟수의 에포크를 지날 때마다 학습률 감소
  - 학습 초기에는 빠른 학습을 진행하다가 전역 최소점 근처에 다다르면 학습률을 줄여 최적점을 찾아갈 수 있도록 해줌
    - optim.lr_scheduler.LambdaLR: 람다 함수의 반환값을 이용해 학습률을 설정함
    - optim.lr_scheduler.StepLR: 일정 step마다 학습률을 gamma 비율만큼 감소시킴
    - optim.lr_scheduler.MultiStepLR: 지정한 여러 epoch에서만 학습률을 gamma 비율만큼 감소시킴
    - optim.lr_scheduler.ExponentialLR: epoch마다 이전 학습률에 gamma를 곱해 감소시킴
    - optim.lr_scheduler.CosineAnnealingLR: 학습률을 코사인 함수 형태로 변화시킴
    - optim.lr_scheduler.ReduceLROnPlateau: 성능 개선 여부에 따라 학습률을 동적으로 감소시킴

````python
import torch
import torchmetrics

# 예측값 (10개의 샘플, 5개 클래스 확률)
preds = torch.randn(10, 5).softmax(dim=-1)
# 실제 정답 (0~4 사이의 정수 10개)
target = torch.randint(5, (10,))

acc = torchmetrics.functional.accuracy(preds, target, task="multiclass", num_classes=5)

print(f"Single Batch Accuracy: {acc}")
````

< 모듈을 이용하여 모델을 평가하는 코드 >

````python
import torch
import torchmetrics

# 모델 평가(정확도) 객체 초기화
metric = torchmetrics.Accuracy(task="multiclass", num_classes=5)

n_batches = 10
for i in range(n_batches):
    # 매 배치마다 가상의 예측값과 정답 생성
    preds = torch.randn(10, 5).softmax(dim=-1)
    target = torch.randint(5, (10,))

    # 현재 배치에 대한 정확도 계산 및 내부 상태 업데이트(데이터 누적)
    acc = metric(preds, target)
    print(f"Accuracy on batch {i}: {acc}")

# 모든 배치(10개)의 데이터를 합산하여 전체 데이터에 대한 최종 정확도 계산
total_acc = metric.compute()
print(f"\nAccuracy on all data: {total_acc}")
````
